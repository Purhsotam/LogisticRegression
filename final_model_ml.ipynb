### 2. Handling Outliers

````python
# Handling Outliers & Outlier treatments
plt.figure(figsize=(25,7))
for i in range(1,len(ml.columns)):
  plt.figure(figsize=(15,4))
  sns.boxplot(ml.iloc[:,i-1])
  plt.plot()
````

We have outlier in fc, px_height and three_g

````python
Q1 = ml['three_g'].quantile(0.25)
Q3 = ml['three_g'].quantile(0.75)
plt.figure(figsize=(15,5))
sns.boxplot(ml['three_g'])
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (ml['three_g'] >= Q1 - 1.5 * IQR) & (ml['three_g'] <= Q3 + 1.5 *IQR)
ml.loc[filter]
plt.figure(figsize=(15,5))
sns.boxplot(ml.loc[filter]['three_g'])
````

````python
Q1 = ml['fc'].quantile(.25)
Q3 = ml['fc'].quantile(.75)
IQR = Q3 - Q1    #IQR is interquartile range. 
plt.figure(figsize=(15,5))
sns.boxplot(ml['fc'])
filter = (ml['fc'] >= Q1 - 1.5 * IQR) & (ml['fc'] <= Q3 + 1.5 *IQR)
ml.loc[filter]
plt.figure(figsize=(15,5))
sns.boxplot(ml.loc[filter]['fc'])
````

````python
Q1 = ml['px_height'].quantile(0.25)
Q3 = ml['px_height'].quantile(0.75)
plt.figure(figsize=(15,5))
sns.boxplot(ml['px_height'])
IQR = Q3 - Q1    #IQR is interquartile range. 

filter = (ml['px_height'] >= Q1 - 1.5 * IQR) & (ml['px_height'] <= Q3 + 1.5 *IQR)
ml.loc[filter]
plt.figure(figsize=(15,5))
sns.boxplot(ml.loc[filter]['px_height'])
````

##### What all outlier treatment techniques have you used and why did you use those techniques?

We have used IQR technique to deal with outlier we have treated the data to replace values within IQR so that outlier will not further effect our data.
